Check if the data frame is correctly imported

### Compare the birthrate and deathrate by years

Import only the relevant datas for visualization to compare the birthrate and deathrate by date and regions. 
To get the average birthrate for entire region by year, select the data for only when region value is equal to 'Whole country'.

```python
#Get the plot of avg birthrate by year(2000-2022)
#access multiple columns with only date,region,birthrate
date = data.loc[:,["Date","Region","Birth","Death"]]
newdate = []
newdate = date.loc[date['Region'] == 'Whole country']
p =pd.DataFrame(newdate)
p

```

Since the original data is saved by each month, calculate the average birth and death rate separately.
Using for loop, save all the average birth rates and death rates in the list.

```python
avgbirthrate= []
avgdeathrate = []
years = ['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
count = 0
sum = 0
for i in range(1,len(newdate)+1):
    count+=1
    sum += float(newdate.loc[18*i,'Birth'])
    if count == 12:
        sum = round(sum/12,3)
        avgbirthrate.append(sum)
        sum = 0
        count = 0
    if i == 270:
        sum = round(sum/6,3)
        avgbirthrate.append(sum)
count= 0
sum = 0 
for i in range(1,len(newdate)+1):
    count+=1
    sum += float(newdate.loc[18*i,'Death'])
    if count == 12:
        sum = round(sum/12,3)
        avgdeathrate.append(sum)
        sum = 0
        count = 0
    if i == 270:
        sum = round(sum/6,3)
        avgdeathrate.append(sum)
df = pd.DataFrame({'Avg Birth Rate': avgbirthrate,
                   'Avg Death Rate': avgdeathrate}, index=years)
df
```

```python
plt.rcParams.update({'font.size': 6}) 
plt.plot(years,avgbirthrate,color = "#539caf",label = "Birth")
plt.plot(years, avgdeathrate,label = "Death",color='maroon')
plt.title("Birth Rate VS Death Rate In South Korea By Year(2000-2022)")
plt.xlabel('Year')
plt.ylabel('Birth and Death')
plt.grid()
plt.legend()
plt.show()
```

Create the bar plot to compare it more visually.

```python
ax = df.plot.bar(rot=0)
```

There are few interesting features in the plot. Comparing the birthrate and deathrate, the birthrate is continuously decreasing since 2000 and deathrate, on the other hand, is increasing. Deathrate starts to outrun birthrate since 2020 and it has approximate difference of 10,000 now.
